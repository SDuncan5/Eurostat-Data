```python
import pandas as pd
import numpy as np
import plotly.express as px
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt
```

Creation of per capita dataset found within Line Plots section of Data Visualization notebook.

```python
df_sd = pd.read_csv("https://raw.githubusercontent.com/SDuncan5/Eurostat-Data/main/sd_eurostat.csv")
df_sd.set_index("Unnamed: 0", inplace=True)
df_sd.index.name = None
df_sd
```

## GDP

```python
df_gdp = pd.read_csv("https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/nama_10_gdp/A.CP_MEUR.B1GQ.BE+BG+CZ+DK+DE+EE+IE+EL+ES+FR+HR+IT+CY+LV+LT+LU+HU+MT+NL+AT+PL+PT+RO+SI+SK+FI+SE+IS+LI+NO+CH+UK+BA+ME+MK+AL+RS+TR+XK+EU27_2020+EU28+EU15+EA+EA20+EA19+EA12/?format=SDMX-CSV&startPeriod=1995&endPeriod=2022&lang=en&label=label_only")
df_gdp
```

```python
df_gdp.rename(columns={'OBS_VALUE': "GDP"}, inplace=True)
df_gdp = df_gdp.drop(["DATAFLOW", "LAST UPDATE", "unit", "na_item", "OBS_FLAG", "freq"], axis=1)
df_gdp
```

## Emigration

```python
df_emigration = pd.read_csv("https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/tps00177/A.TOTAL.COMPLET.TOTAL.NR.T.EU27_2020+EA20+EA19+BE+BG+CZ+DK+DE+EE+IE+EL+ES+FR+HR+IT+CY+LV+LT+LU+HU+MT+NL+AT+PL+PT+RO+SI+SK+FI+SE+IS+LI+NO+CH+UK+ME+MD+MK+AL+RS+TR+UA+BY+RU+SM+KG+TJ+UZ+AM+AZ/?format=SDMX-CSV&returnData=ALL&startPeriod=1990&endPeriod=2021&lang=en&label=label_only")
df_emigration.rename(columns={'OBS_VALUE': "emigration"}, inplace=True)
df_emigration = df_emigration.drop(["DATAFLOW", "LAST UPDATE", "freq", "unit", "OBS_FLAG", "citizen", "agedef", "sex", "OBS_FLAG", "age"], axis=1)
df_emigration
```

```python
px.line(df_emigration, x="TIME_PERIOD", y="emigration", color="geo", markers=True)
```

## Unemployment

```python
df_unemployment = pd.read_csv("https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/tps00203/A.Y15-74.PC_POP.T.EU27_2020+EA20+EA19+BE+BG+CZ+DK+DE+EE+IE+EL+ES+FR+HR+IT+CY+LV+LT+LU+HU+MT+NL+AT+PL+PT+RO+SI+SK+FI+SE+IS+LI+NO+CH+UK+ME+MK+AL+RS+TR/?format=SDMX-CSV&returnData=ALL&startPeriod=2011&endPeriod=2022&lang=en&label=label_only")
# df_unemployment = df_unemployment.drop(["DATAFLOW", "LAST UPDATE", "unit", "OBS_FLAG", "freq", "age"], axis=1)
#pd.DataFrame(~df_unemployment["OBS_VALUE"].isnull())
df_unemployment = df_unemployment[~df_unemployment["OBS_VALUE"].isnull()]
df_unemployment
```

```python
# Filter the DataFrame based on the specified conditions
filtered_df = df_unemployment[
    (df_unemployment["freq"] == "Annual") &
    (df_unemployment["sex"] == "Total") &
    (df_unemployment["unit"] == "Percentage of total population")
]

# Group by 'geo' and 'TIME_VALUE', then calculate the mean of 'OBS_VALUE'
df_unemployment = filtered_df.groupby(['geo', 'TIME_PERIOD'])['OBS_VALUE'].mean().reset_index()
```

```python
# px.line(df_unemployment, x="TIME_PERIOD", y="OBS_VALUE", color="geo", markers=True)
```

```python
df_unemployment.rename(columns={'OBS_VALUE': "unemployment"}, inplace=True)
```

```python
df_unemployment
```

## Population

```python
df_pop = pd.read_csv("https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/demo_gind/A.AVG+CNMIGRAT+CNMIGRATRT+DEATH+FAVG.EU27_2020+EU28+EA20+EA19+BE+BG+CZ+DK+DE+DE_TOT+EE+IE+EL+ES+FR+FX+HR+IT+CY+LV+LT+LU+HU+MT+NL+AT+PL+PT+RO+SI+SK+FI+SE+EEA31+EFTA+IS+LI+NO+CH+UK+BA+ME+MD+MK+AL+RS+TR+UA+XK+GE+AD+BY+MC+RU+SM+AM+AZ/?format=SDMX-CSV&startPeriod=1960&endPeriod=2023&lang=en&label=label_only")
```

```python
df_total_pop = df_pop[df_pop["indic_de"]== "Average population - total"].copy()
df_total_pop.drop(["DATAFLOW", "LAST UPDATE", "OBS_FLAG", "freq", "indic_de"], axis=1, inplace=True)
df_total_pop.rename(columns={'OBS_VALUE': "total_population"}, inplace=True)
df_total_pop
```

```python
df_total_deaths = df_pop[df_pop["indic_de"]== "Deaths - total"].copy()
df_total_deaths.drop(["DATAFLOW", "LAST UPDATE", "OBS_FLAG", "freq", "indic_de"], axis=1, inplace=True)
df_total_deaths.rename(columns={'OBS_VALUE': "total_deaths"}, inplace=True)
df_total_deaths
```

## Compiling data to one dataframe

```python
df_eurostat = df_sd.merge(df_gdp, on=["geo", "TIME_PERIOD"], how="outer")
df_eurostat = df_eurostat.merge(df_emigration, on=["geo", "TIME_PERIOD"], how="outer")
df_eurostat = df_eurostat.merge(df_unemployment, on=["geo", "TIME_PERIOD"], how="outer")
df_eurostat = df_eurostat.merge(df_total_deaths, on=["geo", "TIME_PERIOD"], how="outer")
df_eurostat
```

```python
df_eurostat.to_csv('base_eurostat.csv')
```

```python
df_test = df_eurostat.drop(columns=["Life Satisfaction"])
df_test
```

```python
df_test = df_test[df_test["TIME_PERIOD"] > 2010].dropna()
df_test
```

## Exports, Imports

```python
df_trade = pd.read_csv("https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/tet00002/?format=SDMX-CSV&lang=en&label=label_only")
df_trade
```

```python
df_trade["indic_et"].value_counts()
```

```python
df_total_exports = df_trade[df_trade["indic_et"] == "Exports in million of ECU/EURO"].copy()
df_total_exports
df_total_exports.drop(["DATAFLOW", "LAST UPDATE", "OBS_FLAG", "freq", "indic_et", "sitc06", "partner", "unit"], axis=1, inplace=True)
df_total_exports.rename(columns={'OBS_VALUE': "Exports"}, inplace=True)
df_total_exports
```

```python
df_total_imports = df_trade[df_trade["indic_et"] == "Imports in million of ECU/EURO"].copy()
df_total_imports
df_total_imports.drop(["DATAFLOW", "LAST UPDATE", "OBS_FLAG", "freq", "indic_et", "sitc06", "partner", "unit"], axis=1, inplace=True)
df_total_imports.rename(columns={'OBS_VALUE': "Imports"}, inplace=True)
df_total_imports
```

```python
df_eurostat_trade = df_eurostat.merge(df_total_exports, on=["geo", "TIME_PERIOD"], how="outer")
df_eurostat_trade = df_eurostat_trade.merge(df_total_imports, on=["geo", "TIME_PERIOD"], how="outer")
df_eurostat_trade
```

```python
df_eurostat_trade.to_csv('trade_eurostat.csv')
```

```python
df_test_trade = df_eurostat_trade.drop(columns=["Life Satisfaction"])
df_test_trade
```

```python
df_test_trade = df_test_trade[df_test_trade["TIME_PERIOD"] >= 2010].dropna()
df_test_trade
```
